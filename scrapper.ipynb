```python
from bs4 import BeautifulSoup
from vagalume import find
from time import sleep
import json
import requests
```

```python
url = 'https://www.letras.com/bruno-mars/'
res = requests.get(url)
html_page = res.text

soup = BeautifulSoup(html_page, 'html.parser')
soup.prettify()
```

# Capturando os títulos de algumas músicas

```python
titles = []
for title in soup.find_all('span'):
  titles.append(title.getText())
```

# Temos 199 músicas?

```python
len(titles)
```

```python
data_music = []

for title in titles:
  sleep(1)
  music = find('Bruno Mars', title)
  if music.status == 'exact':
    lirycs = dict()
    lirycs['title'] = music.song.name
    lirycs['liryc'] = music.song.lyric.rsplit('\n')
    data_music.append(lirycs)
```

## Temos apenas 96 músicas

```python
len(data_music)
```

## Salvando o dataset

```python
# Nome do arquivo onde você deseja salvar a lista em formato JSON
nome_arquivo = "data_music.json"

# Abrir o arquivo no modo de escrita
with open(nome_arquivo, 'w') as arquivo:
    # Escrever a lista no arquivo no formato JSON
    json.dump(data_music, arquivo)
```
